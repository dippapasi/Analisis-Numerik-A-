```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 500)
f = np.sin(x) + 0.5*np.sin(2*x) + (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)
```

**Hitung integral dan diferensial dari fungsi f untuk inteval x**

```python
# Turunan
fp = np.gradient(f, x)

# Integral
dx = x[1] - x[0]
F = np.cumsum(f) * dx

print("Turunan -> f'(x)")
print(fp[:10])

print("\nIntegral -> ∫f(x)dx")
print(F[:10])
```

**Output turunan f'(x):** menunjukkan perubahan f terhadap x pada 10 titik pertama, contohnya pada fp(0) = 4.996 sebagai titik pertama. Terlihat hasil turunan untuk titik lainnya sedikit menurun, ini berarti gradien fungsi berkurang seiring x meningkat.

**Output integral F(x):** menunjukkan luas total di bawah kurva f(x) sampai setiap titik, contohnya F(1) = 0.002 yang berarti luas titik pertama sampai titik kedua sebesar 0.002. Terlihat hasil integral selalu bertambah seiring x meningkat.

**Hitung secara terpisah bagian integral di atas dan di bawah y=0**

```python
# Integral kumulatif positif dan negatif
f_pos = np.where(f > 0, f, 0)
f_neg = np.where(f < 0, f, 0)
F_pos = np.cumsum(f_pos) * dx
F_neg = np.cumsum(f_neg) * dx

# Integral total atas dan bawah
I_atas = np.trapezoid(f_pos, x)
I_bawah = np.trapezoid(f_neg, x)


print(f"Integral atas sumbu x (y>0)  = {I_atas:.5f}")
print(f"Integral bawah sumbu x (y<0) = {I_bawah:.5f}")
```

**Output integral atas sumbu x (y>0):** menunjukkan total luas kurva di atas sumbu x sekitar 4.60437 satuan luas. Nilainya selalu positif sebab hanya menjumlah bagian atas sumbu x. Sedangkan **output integral bawah sumbu x (y<0)** menunjukkan total luas kurva di bawah sumbu x (nilainya selalu negatif) sekitar -2.41793. Nilainya selalu negatif sebab bebrada di bawah sumbu x.

**Plot semua fungsi: f, f', F (atas dan bawah dengan area warna berbeda)**

```python
plt.figure(figsize=(10, 12))

# Plot fungsi f(x)
plt.subplot(3, 1, 1)
plt.plot(x, f, color="orange", label="f(x)")
plt.fill_between(x, f, 0, where=(f >= 0), color="purple", alpha=0.4, label="Area Atas (y>0)")
plt.fill_between(x, f, 0, where=(f < 0), color="grey", alpha=0.4, label="Area Bawah (y<0)")
plt.axhline(0, color="black", linewidth=1)
plt.title("Fungsi f(x)")
plt.ylabel("f(x)")
plt.legend()

# Plot turunan f(x)
plt.subplot(3, 1, 2)
plt.plot(x, fp, color="green", label="f'(x)")
plt.axhline(0, color="black", linewidth=1)
plt.title("Turunan f(x)")
plt.ylabel("f'(x)")
plt.legend()

# Plot integral F(x)
plt.subplot(3, 1, 3)
plt.plot(x, F, label="F(x) = ∫ f(x) dx", color="red")
plt.plot(x, F_pos, "--", label="Area positif (f>0)", color="blue")
plt.plot(x, F_neg, "--", label="Area negatif (f<0)", color="purple")
plt.axhline(0, color="black")
plt.title("Integral F(x)")
plt.ylabel("F(x)")
plt.legend()
plt.show()
```

**Plot f(x):** fungsi ditandai dengan garis oranye, area atas (luas positif) berwarna ungu, dan area bawah (luas negatif) berwarna abu-abu. Dengan begitu, bisa terlihat langsung berapa banyak fungsi yang berada di atas maupun di bawah sumbu x.

**Plot f'(x):** fungsi turunan ditandai dengan garis hijau. Titik ketika garis hijau memotong sumbu x biasanya menandai titik maksimum/minimum pada fungsi f(x).

**Plot F(x):** garis merah menunjukkan nilai integral total yang menjumlahkan area positif dengan area negatif sepanjang x. Garis biru menunujukkan akumulasi area positif saja, sedangkan garis ungu menunjukkan akumulasi area negatif saja. Dengan visualisasi memisahkan kedua area tersebut bisa diketahui bahwa integral total (garis merah) merupakan hasil akumulasi dari area positif yang menaikkan nilai integral dan area negatif yang menurunkan nilai integral.
