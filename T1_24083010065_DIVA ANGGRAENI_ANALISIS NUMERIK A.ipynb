```python
# nama:DIVA ANGGRAENI
# NPM: 24083010065

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# range harga jual
x = np.linspace(3000, 10000, 500)

# rescale parameters
a = 6
loc = 3.5 * 1000     # shift loc
scale = 1.5 * 1000   # widen scale

# probability density, right tail skew
y = skewnorm.pdf(x, a=a, loc=loc, scale=scale) * 1000

np.random.seed(65) # masukkan NPM
y = y + np.random.normal(0, 0.005, size=x.shape)  # add Gaussian noise
y = y*1000

# harga produksi
produksi_y= np.linspace(0, 1000, len(x))
produksi_z = np.linspace(3000, 2000, len(x)) + np.random.normal(0, 0.1, size=x.shape)*100
# linearly decreases from 3000 -> 2000 for quantity 0 -> 1000

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y, color="navy", label='penjualan', linewidth=2)
plt.plot(produksi_z, produksi_y, color="red", label='produksi', linewidth=2)
plt.title("Toko Roti Maknyus")
plt.xlabel("x (harga)")
plt.ylabel("y (jumlah barang dibuat/laku)")
plt.grid(True)
plt.legend()
plt.show()

# tugas:
# hitung volume produksi yang menghasilkan kentungan maksimum
#    (pendapatan dari penjualan dikurangi produksi)
# kirim ke muhammad_zulhaj.sada@upnjatim.ac.id
```

**TUGAS**


Hitung volume produksi yang menghasilkan kentungan maksimum (pendapatan dari penjualan dikurangi produksi)

```python
q_vals = produksi_y #barang yang laku
harga_jual = np.interp(q_vals, y, x)
biaya_unit = np.interp(q_vals, produksi_y, produksi_z)

# Pendapatan dan biaya total
pendapatan = harga_jual * q_vals
biaya_total = biaya_unit * q_vals
untung = pendapatan - biaya_total

# Keuntungan maks.
idx = np.argmax(untung)
q_optimal = q_vals[idx]
untung_maks = untung[idx]

print("Produksi optimal:", q_optimal)
print("Keuntungan maksimum:", untung_maks)

# Plot
plt.figure(figsize=(8,5))
plt.plot(q_vals, untung, color="green", label=f"Untung Max={untung.max():.0f}")
plt.axvline(q_optimal, color="red", linestyle="--", label=f"Produksi Optimal={q_optimal:.0f}")
plt.title("Kurva Keuntungan Toko Roti Maknyus")
plt.xlabel("Jumlah produksi (q)")
plt.ylabel("Untung")
plt.legend()
plt.grid(True)
plt.show()
```

Berdasarkan kurva di atas, jumlah produksi roti yang optimal adalah 1000 roti, sehingga keuntungan akan mencapai titik maksimal dengan keuntungan sebesar 7.997.815.
